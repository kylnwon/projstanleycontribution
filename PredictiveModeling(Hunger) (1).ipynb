```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
```

```python
data = pd.read_csv('Downloads/Compiled Annual.csv')
data.info()
print(data)
data.corr()
```

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_absolute_error
from sklearn.feature_selection import f_regression

data.dropna(axis=0, inplace =True)

X = data[[ 'Unempolyment Number ', 'CalFresh Persons ', 'Warn Employees']]
y = data['Under $75000']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 21)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print("P scores for variables calculated from F values, first arrray is F scores for each variable, second array is P scores \
    for each variable. Order is the order in which they appear for X and y variable above.")
print(" ")
print(f_regression(X_train, y_train, center=True))
print(" ")

print(f"R^2 value = {regressor.score(X_test, y_test)}")
print(f"Coefficient values are {regressor.coef_}")
print(f"Intercept value is {regressor.intercept_}")
```
